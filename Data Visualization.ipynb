```python
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
covid=pd.read_csv('/content/COVID-19 Coronavirus.csv')
```

```python
covid.head
```

```python
#Scatter Plot
plt.scatter(covid["Population"], covid["Death percentage"])
plt.title("Country Population and COVID Death Percentage")
plt.xlabel("Population")
plt.ylabel("Death Percent")
plt.show()
```

```python
#Histogram
plt.hist(covid["Death percentage"])
plt.title("Percentages of COVID Deaths by Country")
plt.show()
```

```python
#Line Plot
plt.plot(covid["Population"], covid["Death percentage"])
plt.title("Country Population and COVID Death Percentage")
plt.ylabel("Population")
plt.xlabel("Death Percent")
plt.show()
```

```python
#Box and Whisker
plt.boxplot(covid["Total Cases"])
plt.title("COVID Total case")
plt.show()
```

```python
import plotly.express as px

fig = px.choropleth(covid, locations="ISO 3166-1 alpha-3 CODE",
                    color="Death percentage", # Total is a column of df
                    hover_name="Country", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Reds)
fig.show()
```

```python
sns.violinplot(y=covid['Death percentage'],color='blue')
```

```python
px.bar(covid,'Continent','Total Cases')
```

```python
p=px.density_contour(covid,x='Country',y='Death percentage')
p.show()
```

```python
px.funnel(covid,x='Continent',y='Total Cases')
```

```python
px.scatter(covid,x='Country',y='Death percentage',color='Continent')
```

```python
px.density_heatmap(covid,x='Country',y='Total Cases')
```

```python
#Pie Chart for total cases
px.pie(covid,values='Total Cases',names='Continent')
```
